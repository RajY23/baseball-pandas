```python
import pandas as pd
```

# feature engineering

```python
dfBatting = pd.read_csv('data/baseball-csvs/Batting.csv')
dfBatting.head(2)
```

```python
# BA = HITS / AT BATS
dfBatting['BA'] =  dfBatting['H'] / dfBatting['AB']

# OBP = HITS + BASE BY BALLS + HIT BY PITCH /  (AT BATS + BASE BY BALLS + HIT BY PITCH + SACRAFICED FLY )
dfBatting['OBP'] = (dfBatting['H'] + dfBatting['BB'] + dfBatting['HBP']) /  (dfBatting['AB'] + dfBatting['BB'] + dfBatting['HBP'] + dfBatting['SF'])

# 1B = HITS - sum(2b, 3b, hr)
dfBatting['1B'] = dfBatting['H'] - (dfBatting['2B'] + dfBatting['3B'] + dfBatting['HR'] )

# SLG = 1*1B + 2*2B + 3*3B + 4*HR / AT BATS
dfBatting['SLG'] = ( (1 * dfBatting['1B']) + (2 * dfBatting['2B']) + (3 * dfBatting['3B']) + (4 * dfBatting['HR']) ) / dfBatting['AB']
```

# merge salary data

```python
dfSals = pd.read_csv('data/baseball-csvs/Salaries.csv')
dfSals.head(2)
```

```python
dfBatting = dfBatting[dfBatting.yearID >= 1985]
dfBatting.head(2)
```

```python
mergeddf = pd.merge(dfBatting, dfSals, on=['playerID', 'yearID'])
mergeddf.head(2)
```

```python
mergedf = mergeddf.drop(columns=['G_old', 'teamID_y', 'lgID_y'])
mergedf.head(2)
```

# oakland 2001

```python
oak2001 = mergedf[(mergeddf.yearID == 2001) & (mergedf.teamID_x == 'OAK')]
oak2001.head(2)
```

```python
lostboys = oak2001[oak2001.playerID.isin(['isrinja01', 'giambja01', 'damonjo01', 'saenzol01'])]
lostboys
```

```python
# requirements for new players
lostboys.salary.sum(), lostboys.AB.sum(), lostboys.OBP.mean()
```

# find the right players

```python
players = mergedf[(mergeddf.yearID == 2001) & (mergeddf.teamID_x != 'OAK')][['playerID', 'teamID_x','AB','HR','SLG', 'OBP', 'salary']]
players.head(2)
```

```python
players = players[(players.AB > 10) & (players.salary < 10_000_000)].sort_values(by=['OBP'], ascending=False)
players.head(3)
```
